```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

print('hi this is for github tutorial')

# Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label="Sine Wave")
plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine Wave Graph")
plt.show()
```
